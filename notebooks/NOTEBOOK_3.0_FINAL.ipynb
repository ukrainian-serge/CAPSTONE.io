```python
from bokeh.plotting import figure, output_notebook, show, ColumnDataSource, output_file
from bokeh.models import HoverTool
from bokeh.layouts import column, row

from bokeh.io import curdoc
from bokeh.themes import Theme

from sqlalchemy import create_engine
import pandas as pd
import numpy as np
# output_notebook()

curdoc().theme = Theme(json={'attrs': {
'Figure': {
    'toolbar_location': 'right',
    'outline_line_color': None,
    'min_border_right': 10,
    'sizing_mode': 'stretch_width'
},'Grid': {
    'grid_line_color': None,
},
'Title': {
    'text_font_size': '14pt'
},# apply defaults to Axis properties
'Axis': {
    'minor_tick_out': None,
    'minor_tick_in': None,
    'major_label_text_font_size': '11pt',
    'axis_label_text_font_size': '13pt',
    'axis_label_text_font': 'Work Sans'
    
},
# apply defaults to Legend properties
'Legend': {
    'background_fill_alpha': 0.8,
    'border_line_color' : "black"
}}})

engine = create_engine('postgresql+psycopg2://postgres:postgres@localhost/amazon_reviews')
conn = engine.connect()

query = """SELECT review_date,star_rating FROM books WHERE CAST(review_date AS DATE) > '2010-11-28' AND verified_purchase = 1;"""
book_counts = pd.read_sql_query(query, con=engine, parse_dates='review_date').sample(frac=0.01)
book_counts['day_of_week'] = book_counts.review_date.dt.day_name().str.slice(0, 3)
query = """SELECT review_date,star_rating FROM digital_ebook_purchase WHERE CAST(review_date AS DATE) > '2010-11-28' AND verified_purchase = 1;"""
ebook_counts = pd.read_sql_query(query, con=engine, parse_dates='review_date').sample(frac=0.01)
ebook_counts['day_of_week'] = ebook_counts.review_date.dt.day_name().str.slice(0, 3)
```

```python
from bokeh.transform import jitter
from bokeh.plotting import output_notebook
output_notebook()

DAYS = ['Sun', 'Sat', 'Fri', 'Thu', 'Wed', 'Tue', 'Mon']

source1 = ColumnDataSource(book_counts)
p1 = figure(
#     sizing_mode='stretch_width',
    plot_width=1200, 
    plot_height=400, # for CAPSTONE,io plot_width=700, plot_height=300
    y_range=DAYS, 
    x_axis_type='datetime', 
    tools='', title=None)
p1.circle(x='review_date', 
          y=jitter('day_of_week', width=0.5, range=p1.y_range), 
          alpha=0.5, 
          source=source1,  
          color='black', 
          legend_label='books'
#           size='star_rating'
         )


source2 = ColumnDataSource(ebook_counts)
p2 = figure(
#     sizing_mode='stretch_width',
    plot_width=1200, 
    plot_height=400, # for CAPSTONE,io plot_width=700, plot_height=300
    y_range=DAYS, x_axis_type='datetime', tools='', title=None)
p2.circle(x='review_date', 
          y=jitter('day_of_week', width=0.5, range=p2.y_range),  
          alpha=0.4, 
          source=source2, 
          color='darkorange',
          legend_label='ebooks'
#           size='star_rating'
         )

p1.legend.location = "top_left"
p1.toolbar.logo = None
p2.legend.location = "top_left"
p2.toolbar.logo = None


# output_file('../imgs_charts/1_verified_purchase_reviews.html', 'Verified Purchase Reviews')
show(column([p1, p2]),
#      sizing_mode='stretch_both'
    )
```

*For above charts, a random fractional sample of each format was taken(0.01) because of the size of the data set*  
**Observations:**

<h2 style="text-align:left; font-size:20px;">Rolling average star rating Books vs eBooks</h2>


```python
query = """
SELECT 
    review_date, 
    COUNT(review_id) AS count_reviews,
    ROUND(AVG(star_rating), 4) as avg_stars,
    ROUND(AVG(helpful_votes), 4) as avg_helpful,
    ROUND(AVG(total_votes), 4) as avg_total
    
FROM {} 
GROUP BY review_date 
ORDER BY review_date;
"""
books = pd.read_sql_query(query.format('books'), con=engine, 
                          index_col='review_date', 
                          parse_dates='review_date')

ebooks =pd.read_sql_query(query.format('digital_ebook_purchase'), con=engine, 
                          index_col='review_date',
                          parse_dates='review_date')
```

```python
resampled_books = books.resample('2M').mean()['1996-07-01':]
resampled_ebooks = ebooks.resample('2M').mean()['2007-08-01':]

source_book = ColumnDataSource.from_df(resampled_books)
source_ebook = ColumnDataSource.from_df(resampled_ebooks)

p3 = figure(x_axis_type="datetime",y_range=(3.25, 5.25),
#             sizing_mode='stretch_width',
            plot_width=700, plot_height=500,  # FOR CAPSTONE.io plot_width=600, plot_height=400
            tools="crosshair,pan,wheel_zoom,box_zoom,reset,hover",)
p3.yaxis.axis_label = 'Star Rating'
p3.line(x='review_date', y='avg_stars', source=source_book, color="black", line_width=2.5, line_alpha=.5, legend_label='books')
p3.line(x='review_date', y='avg_stars', source=source_ebook, color="darkorange", line_width=2.5, line_alpha=0.8, legend_label='ebooks')


p3.circle(x='review_date', y='avg_stars', source=source_book, color="black", line_alpha=0.2, legend_label='books')
p3.circle(x='review_date', y='avg_stars', source=source_ebook, color="darkorange", line_alpha=0.5, legend_label='ebooks')

hover = p3.select(dict(type=HoverTool))
hover.tooltips = [("avg_stars", "@avg_stars"), ("review_date", "@review_date{%Y-%b}")]
hover.formatters={'review_date' : 'datetime', 'avg_stars': 'numeral'}
hover.mode = 'mouse'

p3.legend.location = "top_left"
p3.yaxis.major_label_orientation = "vertical"
p3.grid.grid_line_color = None
p3.toolbar.logo = None

# output_file('../imgs_charts/2_month_rolling_avg_star_ratings.html', '2 Month rolling average star rating')
# output_notebook()
# show(p3)

output_notebook()
show(p3)
```

**Observations:**

- It took `Digital` from 2007 - 2012 to stabilize its rolling averages on Amazon Market   
- 2011 - 2012 both formats took a dip
- Rolling average of `Paper` **OUTPERFORMS** `Digital` format starting 2014 - onward

<h2 style="text-align:left; font-size:20px;">Top titles sorted by <b>star rating</b> with stacked <b>helpful review</b> counts for comparison</h2>

```python
query = """SELECT 
	product_title,
	b.star_rating AS book_star, 
	e.star_rating AS ebook_star, 
	b.helpful_votes AS book_helpful, 
	e.helpful_votes AS ebook_helpful, 
	b.total_votes AS book_total, 
	e.total_votes AS ebook_total, 
	b.verified_purchase AS book_verified, 
	e.verified_purchase AS ebook_verified 
FROM (SELECT
		product_title,
		product_id,
		ROUND(AVG(star_rating), 4) AS star_rating,
		SUM(helpful_votes) AS helpful_votes,
		SUM(total_votes) AS total_votes,
		SUM(verified_purchase) AS verified_purchase
	FROM books 
	GROUP BY product_title, product_id
	HAVING COUNT(review_id) > 100 AND SUM(helpful_votes) > 200
) AS b 
JOIN (SELECT
		product_title,
		ROUND(AVG(star_rating), 4) AS star_rating,
		SUM(helpful_votes) AS helpful_votes,
		SUM(total_votes) AS total_votes,
		SUM(verified_purchase) AS verified_purchase
	FROM digital_ebook_purchase 
	GROUP BY product_title, product_id
	HAVING COUNT(review_id) > 100 AND SUM(helpful_votes) > 200
) as e USING(product_title);"""

both = pd.read_sql_query(query, con=engine)
both_grouped = both.groupby('product_title').agg(dict(zip(both.columns.to_list()[1:], [np.mean, np.mean, sum, sum, sum, sum, sum, sum])))
```

```python
from collections import OrderedDict
from bokeh.plotting import figure

### PAPER

df = both_grouped.sort_values("book_star", ascending=False).head(15) # by book stars
titles = [i.title() for i in df.index.values.tolist()]
books_helpful = df.book_helpful.sort_values(ascending=False)
ebooks_helpful = df.ebook_helpful.sort_values(ascending=False)


stars = OrderedDict(titles=[i.title() for i in df.index.values],
                    books_helpful=books_helpful.values, 
                    ebooks_helpful=ebooks_helpful.values)

things = ['books_helpful', 'ebooks_helpful']
colors = ["#c9d9d3", "#718dbf"]
p4 = figure(x_range=stars['titles'], 
#            sizing_mode='stretch_width',
           plot_width=1200, plot_height=400,
           title='Top star rated paper titles',
           toolbar_location='right', tools="pan,box_zoom,hover,reset",
           tooltips=[("Helpful Review Counts", "@$name"),
                     ("Title ", "@titles")]
          )
# [("avg_stars", "@avg_stars"), ("review_date", "@review_date{%Y-%b}")] ]
p4.vbar_stack(things, x='titles', width=0.9, source=stars, 
             color=['black', 'darkorange'], 
             alpha=0.7,
             legend_label=[i.split('_')[0] for i in things]
            )


p4.axis.minor_tick_line_color = None
p4.xaxis.major_label_text_color = None
p4.toolbar.logo = None
p4.legend.title = "helpful review counts for:"
### DIGITAL

df = both_grouped.sort_values("ebook_star", ascending=False).head(15) # by ebook stars
titles = [i.title() for i in df.index.values.tolist()]

books_helpful = df.book_helpful.sort_values(ascending=False)
ebooks_helpful = df.ebook_helpful.sort_values(ascending=False)

stars = OrderedDict(titles=[i.title() for i in df.index.values],
                    books_helpful=books_helpful.values, 
                    ebooks_helpful=ebooks_helpful.values)

things = ['books_helpful', 'ebooks_helpful']
colors = ["#c9d9d3", "#718dbf"]
p5 = figure(x_range=stars['titles'],
#             sizing_mode='stretch_width',
            plot_width=1200, plot_height=400,
            title='Top star rated digital titles',
           toolbar_location='right', tools="pan,box_zoom,hover,reset",
           tooltips=[("Helpful Review Counts", "@$name"),
                     ("Title ", "@titles")]
          )

p5.vbar_stack(things, x='titles', width=0.9, source=stars, 
             color=['black', 'darkorange'], 
             alpha=0.7,
#              legend_label=[i.split('_')[0] for i in things]
            )
            
p5.axis.minor_tick_line_color = None
p5.xaxis.major_label_text_color = None
p5.toolbar.logo = None

### BOTH
# output_file('../imgs_charts/3_helpful_votes_by_stars_both.html', 'Top 30 titles sorted by average star rating')
# show(column([p4, p5]),sizing_mode='stretch_both')
output_notebook()
show(column([p4, p5]))

### SPLIT CHARTS
# output_file('../imgs_charts/3_helpful_reviews_by_stars1.html', 'Top 30 titles by stars')
# show(p4)
# output_file('../imgs_charts/3_helpful_reviews_by_stars2.html', 'Top 30 titles by stars')
# show(p5)
```

**Observation:**

- Above charts top(`Paper`) and bottom(`Digital`) titles are sorted by their respective `star_rating`
  - In all but three cases(`"Fearless", "Words of Radiance", "Being Mortal"`), `Paper` takes the cake in terms of helpful reviews written
